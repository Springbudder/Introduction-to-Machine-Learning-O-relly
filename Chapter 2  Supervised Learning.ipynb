### Supervised learning: 
<p>1. <b>Definition:</b> the machine learning task of inferring a function from labeled training data.</p>
<p>2. <b>Materials:</b> The training data consist of a set of training examples. In supervised learning, each example is a pair consisting of an input object (typically a vector) and a desired output value (also called the supervisory signal). </p>
<p>3. </b>Outcome:</b> A supervised learning algorithm analyzes the training data and produces an inferred function, which can be used for mapping new examples. </p>
<p>4. <b>Utility:</b> An optimal scenario will allow for the algorithm to correctly determine the class labels for unseen instances. This requires the learning algorithm to generalize from the training data to unseen situations in a "reasonable" way.<p>

### In this book: 
<p>1. <b>Utility:</b> supervised learning is used whenever we want to predict a certain outcome from a given input, and we have examples of input/output pairs. </p>
<p>2. <b>Materials:</b> example of input/output pairs data.</p>
<p>3. <b>Goals:</b> Our goal is to make accurate predictions for new, never-before-seen data. </p>
<p>4. <b>Specifics:</b> Supervised learning often requires human effort to build the training set<p>

<hr/>
### Major types of supervised machine learning problems

1. <p> Classfication: the goal is to predict a <b><em>class label.</em></b> </p>
<p>
<ul>
<li> binary classification</li>
<li> multiclass classfication </li>
</ul>
</p>

2. <p> Regression: the goal is to predict a <b><em>continuous number</em></b>.</p>

<hr/>
### Generalization, overfitting and underfitting
<p>1. Generalization</p>
<ul>
<p><b>Definition:</b> If a model is able to make accurate predictions on unseen data, we say it is able to <em>generalize</em> from the training set to the test set.
</ul>

<p>2. Overfitting</p>
<ul>
<p><b>Definition:</b> Building a model that is too complex for the amount of information we have is overfitting.
</ul>

<p>2. Underfitting</p>
<ul>
<p><b>Definition:</b> Model is too simple is underfitting.
</ul>

#### Example: Breast cancer dataset

```python
from sklearn.datasets import load_breast_cancer

# Prepare the data
# 1. load
# 2. split
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify = cancer.target, random_state = 66)

# Compare the influence of different value of neighbors to the model performance
training_accuracy = []
test_accuracy = []
neighbors_settings = range(1, 11)

# Loop the model processing
for neighbor in neighbors_settings:
    clf = KNeighborsClassifier(n_neighbors = neighbor)
    clf.fit(X_train, y_train)
    training_accuracy.append(clf.score(X_train, y_train))
    test_accuracy.append(clf.score(X_test, y_test))
    
# plot the curve of training_accuracy and test_accuracy
plt.plot(neighbors_settings, training_accuracy, label = "training accuracy")
plt.plot(neighbors_settings, test_accuracy, label = "testing accuracy")
plt.ylabel("Accuracy")
plt.xlabel("value of n_neighbors")
plt.legend()
```

#### Ridge regression

```python
from sklearn.linear_model import Ridge

ridge = Ridge().fit(X_train, y_train)
print "Training set score: {:.2f}".format(ridge.score(X_train, y_train))
print "Test set score: {:.2f}".format(ridge.score(X_test, y_test))
```

#### Note:
<p> very important parameter in ridge regression: <b><em>$\alpha$</em></b> </p>
<ul>
<li> high: penalty much, coefs were condensed harder, <b>more restricted model</b></li>
<li> low: penalty less, coefs were condensed softer </li>
</ul>

```python
ridge10 = Ridge(alpha=10).fit(X_train, y_train)
print("Training set score: {:.2f}".format(ridge10.score(X_train, y_train)))
print("Test set score: {:.2f}".format(ridge10.score(X_test, y_test)))
```

```python
ridge01 = Ridge(alpha=0.1).fit(X_train, y_train)
print("Training set score: {:.2f}".format(ridge01.score(X_train, y_train)))
print("Test set score: {:.2f}".format(ridge01.score(X_test, y_test)))
```

#### We can see $\alpha\ =\ 0.1$ is the best in $\alpha\ =\ [\ 0.1,\ 1,\ 10\ ]$

```python
plt.rcParams["figure.figsize"] = (20,6)
plt.plot(ridge.coef_, 's', label="Ridge alpha=1")
plt.plot(ridge10.coef_, '^', label="Ridge alpha=10")
plt.plot(ridge01.coef_, 'v', label="Ridge alpha=0.1")
plt.plot(lr.coef_, 'o', label="LinearRegression")
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitude")
plt.hlines(0, 0, len(lr.coef_))
plt.ylim(-25, 25)
plt.legend()
plt.rcParams["figure.figsize"] = (8.0, 6.0)
```

### Decision Tree

```python
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=42)

tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train)

print("Accuracy on training set: {:.3f}".format(tree.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(tree.score(X_test, y_test)))
```

####  Control the complexity of tree: pre-pruning and pro-pruning
<p> Example: <b>pre-pruning</b> </p>
<p>Critical parameter: <b><em>max_depth</em></b></p>

```python
pre_pruning_tree = DecisionTreeClassifier(max_depth=4, random_state=0)
pre_pruning_tree.fit(X_train, y_train)
print("Accuracy on pre-pruning training set: {:.3f}".format(pre_pruning_tree.score(X_train, y_train)))
print("Accuracy on pre-pruning test set: {:.3f}".format(pre_pruning_tree.score(X_test, y_test)))
```

```python
plt.rcParams["figure.figsize"] = (15, 4)
def plot_feature_importances_cancer(model):
    n_features = cancer.data.shape[1]
    plt.bar(range(n_features), model.feature_importances_)
    plt.xticks(np.arange(n_features), cancer.feature_names, rotation = 70)
    plt.xlabel("Feature")
    plt.ylabel("Feature importance")
    
plot_feature_importances_cancer(tree)
plt.rcParams["figure.figsize"] = (8, 6)
```

#### Ensembles of Decision Trees

+ Definition of Ensemble: combine multiple models to a more powerful model

A main drawback of decision trees is they tend to <strong><em>overfit the data</em></strong> $\ \Rightarrow\ $ <strong>Random Forest</strong>

> A random forest is essentially a collection of decision trees, where each tree is slightly different from the others. The idea behind random forest is that each tree might do a relatively good job of predicting, but will likely overfitting on part of the data. If we build many trees, all of which work well and overfit in different ways, we can reduce the amount of overfitting by averaging their results. 

+ Critical parameters:
    + <b><em>n_estimators</em></b>
    + <b><em>max_features</em></b>
    + <b><em>max_depth</em></b>
    + <b><em>n_jobs</em></b>
$$\ $$
+ Advantages:
    + Easy to understand
    + Do not need to scaling
    + Without heavy tuning of the parameters
    + Can generalize
$$\ $$   
+ Disadvantages
    + Do not tend to perform well on very high dimensional, sparse data, such as text data (Linear model is better)
    + Time and memory consuming
<hr/>

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_moons

X, y = make_moons(n_samples = 100, noise = 0.25, random_state = 3)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, random_state = 42)

forest = RandomForestClassifier(n_estimators = 5, random_state = 2)
forest.fit(X_train, y_train)

print "Random Forest training data set score: {:.4f}".format(forest.score(X_train, y_train))
print "Random Forest testing data set score: {:.4f}".format(forest.score(X_test, y_test))
```

```python
fig, axes = plt.subplots(2, 3, figsize = (20, 10))

for i, (ax, tree) in enumerate(zip(axes.ravel(), forest.estimators_)):
    ax.set_title("Tree {}".format(i))
    mglearn.plots.plot_tree_partition(X_train, y_train, tree, ax = ax)
    
mglearn.plots.plot_2d_separator(forest, X_train, fill = True, ax = axes[-1, -1], alpha = 0.4)
axes[-1, -1].set_title("Random Forest")
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
```

#### Kernel tricks

In order to differentiate the data, we need some other fabricated dimension, but which combination or transforming is good for us? Here we use some tricks:

+ The plolynomial kernel
+ The radial basis function (RBF, aka Gaussian kernel)
    + Infinite-dimensional feature space
    + Considered all possible polynomials of all degrees
    + Importance of the features decreases for higher degrees(Tylor expansion)
    
<p> Distances between points in Gaussian kernels: </p>
    $$k_{rbf}(x_{1}, x_{2})\ =\ exp(\ \gamma||x_{1} - x_{2}||^2)$$

#### Example of RBF kernel SVM

```python
from sklearn.svm import SVC

X, y = mglearn.tools.make_handcrafted_dataset()

svm = SVC(kernel = 'rbf', C = 10, gamma = 0.1).fit(X, y)
mglearn.plots.plot_2d_separator(svm, X, eps = 0.5)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)

sv = svm.support_vectors_

sv_labels = svm.dual_coef_.ravel() > 0
mglearn.discrete_scatter(sv[:, 0], sv[:, 1], sv_labels, s = 15, markeredgewidth = 3)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

#### Key parameters in SVM (RBF kernel):
+ $C\ $: Regularization parameters, samll means influence of every points is limited, so the model was restricted, boundary tend to be linear
+ $\gamma\ $: Width of Gaussian kernel, small means the distances are determined far for many other points, so the boudary will be more no-linear

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state = 0)

svc = SVC()

svc.fit(X_train, y_train)

print "Accuracy on training set: {:.2f}".format(svc.score(X_train, y_train))
print "Accuracy on testing set: {:.2f}".format(svc.score(X_test, y_test))
```

#### Why overfitting?

<p> Because SVM is very sensitive for parameters. </p>

#### How to solve it?

<p> Just standardized it (scale it).

```python
min_on_training = X_train.min(axis = 0)
range_on_training = (X_train - min_on_training).max(axis = 0)

X_train_scaled = (X_train - min_on_training) / range_on_training
X_test_scaled = (X_test - min_on_training) / range_on_training
print "Minimum for each feature\n{}".format(X_train_scaled.min(axis = 0))
print "Maximum for each feature\n {}".format(X_train_scaled.max(axis = 0))
```

```python
svc = SVC()
svc.fit(X_train_scaled, y_train)

print "Accuracy on training set: {:.3f}".format(svc.score(X_train_scaled, y_train))
print "Accuracy on test set: {:.3f}".format(svc.score(X_test_scaled, y_test))
```

#### Underfitting!!! 

#### Tuning $\ C\ $ and $\ \gamma\ $

```python
svc = SVC(C = 1000)
svc.fit(X_train_scaled, y_train)

print "Accuracy on training set: {:.3f}".format(svc.score(X_train_scaled, y_train))
print "Accuracy on test set: {:.3f}".format(svc.score(X_test_scaled, y_test))
```

#### Example of MLPs in two_moons dataset

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split

X, y = make_moons(n_samples = 100, noise = 0.25, random_state = 3)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, random_state = 42)
mlp = MLPClassifier(solver = 'lbfgs', random_state = 0).fit(X_train, y_train)

mglearn.plots.plot_2d_separator(mlp, X_train, fill = True, alpha = 0.3)
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")

print "Accuracy on the training dataset: {:.3f}".format(mlp.score(X_train, y_train))
print "Accuracy on the testing dataset: {:.3f}".format(mlp.score(X_test, y_test))
```

#### A very important parameter is n_nodes of hidden layer, the default setting is 100, which is too much for the present dataset

```python
mlp = MLPClassifier(solver = 'lbfgs', random_state=0, hidden_layer_sizes=[10])
mlp.fit(X_train, y_train)
mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")

print "Accuracy on the training dataset: {:.3f}".format(mlp.score(X_train, y_train))
print "Accuracy on the testing dataset: {:.3f}".format(mlp.score(X_test, y_test))
```

<p>Seems perform similarly, except the ragged boudary, since we used the default "relu" for nonlineaty.</p>
#### Change parameters: 2 hidden layers, 10 units (nodes) each

```python
mlp_two_hidden_layers = MLPClassifier(solver = 'lbfgs', activation = "tanh", random_state = 0, hidden_layer_sizes = [10, 10])

mlp_two_hidden_layers.fit(X_train, y_train)
mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
print "Accuracy on the training dataset: {:.3f}".format(mlp_two_hidden_layers.score(X_train, y_train))
print "Accuracy on the testing dataset: {:.3f}".format(mlp_two_hidden_layers.score(X_test, y_test))
```

#### Another very important parameter: $\ \alpha\ $: Like the L2 regularization

+ <b>high:</b> penalty is much, so effects of each data points is limited, the model tends to be linear
+ <b>low:</b> penalty is less, tends to be non-linear

```python
fig, axes = plt.subplots(2, 4, figsize = (20, 8))

for axx, n_hidden_nodes in zip(axes, [10, 100]):
    for ax, alpha in zip(axx, [0.0001, 0.01, 0.1, 1]):
        mlp = MLPClassifier(solver = 'lbfgs', random_state = 0, \
                            hidden_layer_sizes = [n_hidden_nodes, n_hidden_nodes],\
                            alpha = alpha)
        mlp.fit(X_train, y_train)
        mglearn.plots.plot_2d_separator(mlp, X_train, fill = True, alpha = 0.3, ax = ax)
        mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train, ax = ax)
        ax.set_title("n_hiddens = [{}, {}]\nalpha = {:.4f}".format(n_hidden_nodes, n_hidden_nodes, alpha))
```

### Now we need to summarize the important parameters for Neural Network Moldes (Multi-Layers Perceptrons)

+ <b><em>hidden_layer_sized:</em></b> [number_of_layers, nodes_of_each_layer]
+ <b><em>$\alpha\ $:</em></b>$\ $regularization parameter
    + low: non_linear
    + high: linear
+ <b><em>solver: </em></b> Usally we use "lbfgs" or "Adam", deep learning uses "sgd".
+ <b><em>random_state: </em></b> different random seeds generates different models

#### Example of breast cancer dataset

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify = cancer.target, random_state = 0)

mlp = MLPClassifier(random_state = 0).fit(X_train, y_train)

print "Accuracy on the trainning dataset: {:.4f}".format(mlp.score(X_train, y_train))
print "Accuracy on the testing dataset: {:.4f}".format(mlp.score(X_test, y_test))
```

The result is obviously underfitting, what can we do?

#### Scaling!

Uniform:$$X_{scaled} \ =\ \frac{X_{max} - X_{min}}{X_{max}}$$

Standardization: $$X_{scaled}\ =\ \frac{X - E(X)}{std(X)}$$

Here we use the later one: <b>Standardization</b>

```python
X_train_scaled = (X_train - X_train.mean(axis = 0)) / X_train.std(axis = 0)
X_test_scaled = (X_test - X_train.mean(axis = 0)) / X_train.std(axis = 0)

mlp = MLPClassifier(max_iter = 1000, random_state = 0, alpha = 1).fit(X_train_scaled, y_train)

print "Accuracy on the trainning dataset: {:.4f}".format(mlp.score(X_train_scaled, y_train))
print "Accuracy on the testing dataset: {:.4f}".format(mlp.score(X_test_scaled, y_test))
```
