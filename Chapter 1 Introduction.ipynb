````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_dataset["data"], iris_dataset["target"], random_state = 0)
````

### K-Nearest neighbors
------------
1. classification algorithms
2. The k in this algorithm means the k nearest points counted as a catagory
3. KNeighborsClassifier (n_neighbors = ?)

````python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1)
````

````python
knn
````

````python
knn.fit(X_train, y_train)
````

````python
knn
````

````python
X_new = np.array([[5, 2.9, 1, 0.2]])
print "X_new.shape:: {}".format(X_new.shape)
````

### Evaluating the model

1. Accuracy: right item predicted
```
score = classifier.score(test_input, test_output)
```

````python
# Here we use the package function -------> Classifier.score(fetures, result)
print "\n"
print "Test score used the package funcion ----->Classifier.score() is: {:>34}".format(knn.score(X_test, y_test))
print "\n"

# Here we make our own score-calculating process
y_prediction = knn.predict(X_test)
print "Test score use our own process is: {:>64}".format(np.mean(y_prediction == y_test))
print "\n"
````
